```python
from hdf5storage import loadmat
import numpy as np
from os.path import join
```

```python
num_img = 350
ds_path = 'D:/datasets/processed/voc2012'
rgb_path = join(ds_path, 'rgb/val')
gt_path = join(ds_path, 'truth/val')
lgt_path = join(ds_path, 'deeplab_prediction/val')

rgb_fmt = 'val_%06d_rgb.jpg'

gt_fmt = 'val_%06d_pixeltruth.mat'
gt_name = 'truth_img'

lgt_fmt = 'val_%06d_logits.mat'
lgt_name = 'logits_img'
```

```python
num_wrong = np.zeros((num_img), dtype=np.uint64)
```

```python
for idx in range(1, num_img+1):
    logits = loadmat(join(lgt_path, lgt_fmt % idx))[lgt_name][...,1:]
    gt = loadmat(join(gt_path, gt_fmt % idx))[gt_name].ravel()
    
    pred = np.argmax(logits, axis=-1).ravel()+1
    
    mask = (gt>0) & (gt<255)
    gt = gt[mask]
    pred = pred[mask]
    
    num_wrong[idx-1] = (pred!=gt).sum()
```

```python
most_wrong = list(reversed(np.argsort(num_wrong)+1))
```

```python
with open('most_wrong.txt', 'w') as f:
    f.write('\n'.join([str(idx) for idx in most_wrong]))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from skimage.io import imread
```

```python
def show_rgbgtp(idx):
    global class_labels
    
    im = imread(join(rgb_path, rgb_fmt % idx))
    gt = loadmat(join(gt_path, gt_fmt % idx))[gt_name]
    lgts = loadmat(join(lgt_path, lgt_fmt % idx))[lgt_name][...,1:]
    
    p = np.argmax(lgts, axis=-1)+1
    p[(gt==0) | (gt==255)] = 0
    
    gt_comm = dict()
    for lab in np.unique(gt):
        gt_comm[class_labels[min(lab, len(class_labels)-1)]] = np.sum(gt==lab)
        
    pred_comm = dict()
    for lab in np.unique(p):
        pred_comm[class_labels[lab]] = np.sum(p==lab)
        
    print('GT:', gt_comm)
    print('P:', pred_comm)
    
    plt.imshow(im)
    plt.show()
    
    plt.imshow(gt, 'jet', vmin=0, vmax=21)
    plt.show()
    
    plt.imshow(p, 'jet', vmin=1, vmax=21)
    plt.show()
    
    errors = (p!=gt)
    errors[(gt==0) | (gt==255)] = 0
    plt.imshow(errors, 'gray')
    plt.show()
```

```python
show_rgbgtp(226)
```
