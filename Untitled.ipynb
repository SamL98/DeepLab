```python
from hdf5storage import loadmat
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
gt = loadmat('D:/datasets/processed/voc2012/truth/val/val_%06d_pixeltruth.mat' % 8)['truth_img']
```

```python
plt.imshow(gt, 'jet')
plt.show()
```

```python
fg = np.ones_like(gt)
fg[(gt==0) | (gt==255)] = 0
#fg[gt==255] = 0
```

```python
from scipy.ndimage.morphology import distance_transform_edt as dt
```

```python
wmap = dt(fg)
```

```python
wmap = 1 - (1./np.maximum(1, wmap))**2
```

```python
plt.imshow(fg, 'gray')
plt.show()
plt.imshow(wmap, 'gray')
plt.show()
```

```python
print(wmap.min(), wmap.mean(), wmap.max())
```

```python
hist = np.histogram(wmap.ravel(), bins=100)[0]
```

```python
plt.bar(range(100), hist)
plt.show()
```

```python
wmap_masked = wmap.copy()
wmap_masked[wmap<np.percentile()] = 0
wmap_masked.mean()
```
